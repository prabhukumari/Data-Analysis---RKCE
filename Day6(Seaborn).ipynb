### Distribution plot

- univariate distribution

```python
help(sns.distplot)
```

```python
sns.distplot(df["petal_length"])
```

```python
sns.distplot(df["petal_length"],kde=False)
```

```python
sns.distplot(df["petal_length"],hist=False,color="g")
```

### Bivariate Distribution
- sns.jointplot()

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df)
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind="hex")
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind="hist")
```

### pairplot
- sns.pairplot()

```python
help(sns.pairplot)
```

```python
sns.pairplot(df,hue="species",kind='scatter',diag_kind='hist')
```

```python
sns.pairplot(df,hue="species",kind='scatter',diag_kind='hist',palette='husl')
```
